```python
import numpy as np  # NumPy untuk komputasi numerik
import matplotlib.pyplot as plt  # Matplotlib untuk plotting
```

```python
# Data training untuk model harga rumah
x_train = np.array([1.0, 2.0, 4.0])  # Luas rumah dalam meter persegi (1000m^2)
y_train = np.array([300.0, 500.0, 800.0])  # Harga rumah dalam jutaan rupiah (Rp)
```

```python
plt.scatter(x_train, y_train)  # Membuat scatter plot dengan x_train sebagai sumbu x dan y_train sebagai sumbu y
plt.show()  # Menampilkan plot
```

```python
w=180
b=100

# Mendapatkan jumlah total sampel dalam data latih
m= x_train.shape[0]

# Variabel untuk menyimpan total biaya
cost_sum = 0

for i in range(m):
  # Menghitung harga yang diprediksi untuk rumah ke-i menggunakan model
  f_wb = w * x_train[i] + b
  cost = (f_wb - y_train[i])**2
  cost_sum = cost_sum + cost
  total_cost = (1/(2*m))*cost_sum
total_cost
```

```python
def compute_cost(x,y,w,b):
  m=x.shape[0]
  cost_sum = 0
  for i in range(m):
    f_wb= w * x[i] + b
    cost = (f_wb - y[i])**2
    cost_sum = cost_sum + cost
  total_cost = (1/(2*m))*cost_sum

  return total_cost
```

```python
compute_cost(x_train,y_train,180,100)
```

```python
compute_cost(x_train,y_train,100,100)
```
